# Window Generation
This notebook is for generating and visualizing the Hann window for the GPR, which includes the predistortion for the GPR.

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create a hann window
npts = 1024

window = np.hanning(npts)
fig1 = plt.figure()
plt.title("Radar Hann Window")
plt.xlabel("Sample (Location in RAM)")
plt.ylabel("Amplitude Scale Factor")
plt.plot(window)
```

```python
#write the hann window to the c header file
filepath = "D:/HSL/hs2-gpr/Core/Inc/"

#start the file
f = open(filepath + "hann.h", "w+")
f.write("#ifndef INC_HANN_H_\n#define INC_HANN_H_\n\n\n")
f.write("uint16_t window[1024] = {\n")

for i in range(npts):
    f.write("\t\t")
    f.write(str(int(window[i] * (2**14 - 1))))
    f.write(",\n")

#end the file
f.write("};\n\n")
f.write("#endif /* INC_HANN_H_ */")
f.close()
```
